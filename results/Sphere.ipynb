```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A2 = lambda th : np.array([
    [0,0,0],
    [0,np.cos(th),-np.sin(th)],
    [0,np.sin(th),np.cos(th)]
])

B = lambda th : np.array([
    [np.cos(th),-np.sin(th),0],
    [np.sin(th),np.cos(th),0],
    [0,0,1]
])

def Calc_points(thA,thB):
  A_case = A2(thA * np.pi)
  B_case = B(thB * np.pi)

  ActualA = np.eye(3)
  ActualB = np.eye(3)

  n,m = int(np.ceil(2 / thA)), int(np.ceil(2 / thB))
  n = 1 * n
  m = 1 * m

  output = np.zeros((n,m,3))
  st = np.zeros(3)
  st[1] = 1

  for i in range(n):
    for j in range(m):
      output[i,j,:] = ActualB @ ActualA @ st
      ActualB = ActualB @ B_case
    ActualA = ActualA @ A_case

  return output
```

```python
from matplotlib import cm
import matplotlib.colors as mcolors
import matplotlib.colorbar as mcolorbar
from matplotlib.colors import Normalize
```

```python
data = Calc_points(10/101, 4/ 101)

n,m,_ = data.shape
data = data.reshape(-1,3)
colors = np.array([[cm.inferno(i / n) for _ in range(m)] for i in range(n)]).reshape(-1,4)

fig = plt.figure(figsize = (5,5))
ax = fig.add_subplot(projection = "3d")

im = ax.scatter(data[:,0],data[:,1],data[:,2],color = colors,alpha = 0.3,s = 5)


norm = Normalize(vmin=0, vmax=n)
cbar = plt.colorbar(cm.ScalarMappable(norm=norm, cmap='inferno'), ax=ax, fraction=0.03, pad=0.15)
cbar.set_label('j', rotation=0, labelpad=15)

ax.xaxis.pane.set_color((1.0, 1.0, 1.0, 0.0))
ax.yaxis.pane.set_color((1.0, 1.0, 1.0, 0.0))
ax.zaxis.pane.set_color((1.0, 1.0, 1.0, 0.0))

ax.set_xlabel(r"$\langle 1 | j,k\rangle$")
ax.set_ylabel(r"$\langle 2 |j,k\rangle$")
ax.set_zlabel(r"$\langle 3 |j,k\rangle$")
plt.show()
```
